**Introduction:**

The purpose of this project is to predict the popularity of a song on Spotify. This prediction can be used to inform music marketing decisions and help record labels to select songs with the highest chances of success. In order to accomplish this, a variety of machine learning algorithms were used to analyze the data. The algorithms used include Extra Tree Regressor, Linear Regression, Random Forest Regressor, KNN Regressor, and XGBoost Regressor. All of these algorithms were used to analyze the data and make predictions on the popularity of a song.



**Abstract:**

This project aims to predict the popularity of songs on Spotify using machine learning algorithms. The project uses Extra tree regressor to identify the top features that are responsible for the popularity of the songs. Subsequently, linear regression, Random Forest regressor, KNN regressor, and XGBoost regressor are applied to those top features. Mean squared error is then calculated and compared between the results from the models. The results show that KNN with n=5 provides the closest results to the actual popularity values. The project then provides a brief overview of the algorithms used.

**Dataset:**

The datasets used in this project were obtained from Kaggle and include various metrics related to the popularity of a song on Spotify. The metrics include the songâ€™s artist, title, release date, genre, tempo, duration, loudness, energy, and the number of streams. **Reading the Spotify dataset from kaggle **

```python
df = pd.read_csv('/kaggle/input/-spotify-tracks-dataset/dataset.csv')
df.head(5)
```

**Removing unnecessary cells**

```python
df = df.drop(['track_id','track_name'],axis=1)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

**Dividing the X values and Y values, so that we can train and test the model**

```python
X = df.drop(['popularity'],axis=1)
Y = df['popularity']
```

```python
X
```

**Pre-processing of data (encoding of the unreadable data)**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
X['genre_label']=le.fit_transform(X['track_genre'])
```

```python
X.drop(['track_genre'],axis=1)
```

**Encoding the data with string/Object values**

```python
le = LabelEncoder()
X['artists_label']=le.fit_transform(X['artists'])
X=X.drop(['artists','album_name','explicit','track_genre'],axis=1)
```

```python
X.info()
```

**Splitting the dataset into training and test models**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
x_train
```

**Linear Regression**

Linear Regression is a supervised machine learning algorithm that is used for predicting the popularity of a song. It is a linear approach that models the relationship between a dependent variable and one or more independent variables. The model is used to find the best fit line that describes the relationship between the dependent and independent variables. The popularity of the song is then predicted using this best fit line.

**Implimenting Linear Regression on training model**

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
pred = lm.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,pred))
```

```python
import seaborn as sns
sns.scatterplot(y_test,pred)
```

```python
df_test = x_test
df_test['popularity'] = y_test
df_test['linear_pred'] = pred
df_test
```

**Hyperparameter Tunning using Extra Trees Regressor**

**why hyper tunning the model?**

Hyperparameter tuning is an important part of the machine learning process. It helps to find the best parameters to optimize the performance of a model. Tuning the hyperparameters can significantly improve the accuracy and robustness of a model, and it can also help to reduce overfitting and underfitting. Hyperparameter tuning is also necessary to ensure that a model is not overfitting the training data and generalizing well to unseen data. Hyperparameter tuning is often done using a combination of manual and automated techniques.

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model = ExtraTreesRegressor()
model.fit(x_train,y_train)
```

**Extra Trees Regressor**

The Extra Tree Regressor is a machine learning algorithm that is used to predict the popularity of a song. It is an ensemble learning method that uses decision tree predictors to suggest the most important factors that affect the popularity of a song. It uses a randomized approach to identify the optimal split points of a set of data and then builds multiple decision trees by assigning each data point to a different tree. This allows the model to identify the most important features that contribute to the popularity of the song.

```python
model.feature_importances_
```

```python
importances = pd.DataFrame({
    'feature':X.columns,
    'importances':model.feature_importances_
}).sort_values('importances',ascending=False)
```

**Data visualization of retreived Important features responsible for popularity of music**

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,20))
plt.title('Important Features')
sns.barplot(data=importances,x="importances",y="feature")
```

**Considering the top important features responsible for popularity**
(genre_label,artists_label,accoustiness,valence,danceability,energy, loudness,instrumentalness)

**Dividng X-train and X-test values**

```python
x_train = x_train[['genre_label','artists_label','acousticness','valence','danceability','energy', 'loudness','instrumentalness']]
x_train.head()
```

```python
x_test = x_test[['genre_label','artists_label','acousticness','valence','danceability','energy', 'loudness','instrumentalness']]
x_test.head()
```

**RandomForestRegressor**
(we fit Random Forest model on the important features)  

The Random Forest Regressor is an ensemble method that is used to predict the popularity of a song. It is an ensemble learning technique that combines multiple decision tree models to form an overall prediction. Each decision tree model is fit on a random subset of the data, and the output of each decision tree model is then combined to form the final prediction. The Random Forest regressor is useful in cases where the data is too complex for a single decision tree model.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(100,1200,12)]
max_features = ['auto','sqrt']
max_depth = [int(x) for x in np.linspace(5,30,6)]
min_samples_split = [2,5,10,15,100]
min_samples_leaf = [1,2,5,10]
```

```python
rf = RandomForestRegressor()
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
random_rf = RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring="neg_mean_squared_error",n_iter=5,cv=5,random_state=42,verbose=2)
```

```python
random_rf.fit(x_train,y_train)
```

```python
print(random_rf.best_params_)
```

```python
print(random_rf.best_score_)
```

```python
pred=random_rf.predict(x_test)
```

```python
sns.distplot(y_test-pred)
```

```python
df_test['rf'] = pred
```

```python
df_test.head(10)
```

**Implementing KNN regressor**

**KNN Regressor**

KNN Regressor is a machine learning algorithm that is used to predict the popularity of a song. It is a supervised learning algorithm that uses the k-nearest neighbors algorithm to identify the most similar songs and then uses the average popularity of those songs as the predicted popularity of the song. The value of k is usually chosen to be 5, which means that the algorithm looks at the 5 most similar songs and their average popularity to predict the popularity of the song. 

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k = KNeighborsRegressor(n_neighbors=3)
```

```python
k.fit(x_train,y_train)
```

**calculating Mean Squared error for n=1,3,5,7 values**

```python
mse = []
n = [1,3,5,7,9]
for i in n:
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred = knn.predict(x_test)
    mse.append(mean_squared_error(y_test,pred))
mse
```

**MSE value is less for n=5**

**prediction using KNN**

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train,y_train)
pred = knn.predict(x_test)
```

```python
df_test['knn']=pred
```

```python
df_test
```

```python
df_test=df_test[['genre_label','artists_label','acousticness','valence','danceability','energy', 'loudness','instrumentalness','popularity','linear_pred','rf','knn']]
```

**Implementing XGBoost Regressor**

** XGBoost Regressor: **

XGBoost Regressor is an ensemble machine learning algorithm that is used to predict the popularity of a song. It is an optimized version of the gradient boosting algorithm that uses tree-based models to find the best fit for the data. The algorithm uses a gradient-descent based optimization algorithm to find the best fit for the data, and then uses the model to predict the popularity of the song.


```python
from xgboost import XGBRegressor
```

```python
XGBreg = XGBRegressor(scale_pos_weight=1,
                      learning_rate=0.01,  
                      colsample_bytree = 0.3,
                      subsample = 0.8,
                      n_estimators=1000, 
                      reg_alpha = 0.3,
                      max_depth=10, 
                      gamma=1)
```

```python
XGBreg.fit(x_train,y_train)
pred = XGBreg.predict(x_test)
```

```python
print(mean_squared_error(y_test,pred))
```

```python
df_test['XGBreg']=pred
```
